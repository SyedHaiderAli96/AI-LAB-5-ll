```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Provided data
areas = [2600, 3000, 3200, 3600, 4000]
prices = [550000, 565000, 610000, 680000, 725000]

# Create a DataFrame from the data
df = pd.DataFrame({'area': areas, 'price': prices})

# Visualizing the data
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.title('Area vs Price')
plt.show()

# Prepare data for training
X = df[['area']]
y = df['price']

# Create linear regression model
model = LinearRegression()
model.fit(X, y)

# Predicting prices for the given areas in the dataset
predictions = model.predict(X)
df['predicted_price'] = predictions
print(df)

# Displaying the coefficients
print(f"Coefficient (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Example prediction (3300 sq ft)
example_area = 3300
predicted_price = model.predict([[example_area]])
print(f"Predicted price for a home with {example_area} sq ft area: {predicted_price[0]}")
```
