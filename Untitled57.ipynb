```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Simulating a dataset for demonstration
data = {
    'area': [1200, 1500, 1800, 2000, 2500],
    'price': [240000, 300000, 360000, 400000, 500000]
}
df = pd.DataFrame(data)

# Visualizing the dataset
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.title('Area vs Price (New Dataset)')
plt.show()

# Prepare data for training
X = df[['area']]
y = df['price']

# Create linear regression model
model = LinearRegression()
model.fit(X, y)

# Predicting prices for the areas in the dataset
predictions = model.predict(X)
df['predicted_price'] = predictions
print(df)

# Displaying the coefficients
print(f"Coefficient (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Example prediction (predict price for 3300 sq ft)
example_area = 3300
predicted_price = model.predict([[example_area]])
print(f"Predicted price for a home with {example_area} sq ft area: {predicted_price[0]}")
```
